```python
import os
import base64
from io import BytesIO

from PIL import Image

from mapboxgl.slippy_map import TileServer
from mapboxgl.utils import *
from mapboxgl.viz import *
```

```python
token = os.getenv('MAPBOX_ACCESS_TOKEN')
```

```python
file = 'https://s3-us-west-2.amazonaws.com/remotepixel/data/images/LC08_L1TP_016034_20180120_20180206_01_T1_cog.tif'
ts = TileServer(file, nodata=0, tile_size=512, tile_format='jpg', port=8081)
ts.stop()
ts.start()
tiles_url = ts.get_url()
print(tiles_url)
```

```python
viz = RasterTilesViz(tiles_url, 
                     tiles_size=512,
                     tiles_bounds=ts.get_bounds(),
                     center=ts.get_center(),
                     access_token=token, 
                     height='1000px', 
                     zoom=13)
viz.show()
```
